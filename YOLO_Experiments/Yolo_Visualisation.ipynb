```python
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
import cv2
import data_utils
```

```python
# Opening a model that was alraedy trained & fine-tuned

path_to_trained_model_weights = 'project_folder/train5/weights/best.pt'

model = YOLO(path_to_trained_model_weights)
```

```python
# Checking out results on a folder of images (train)

results_train = model('/Users/vihaan/Workspace/Local_AI_Earth/DataManipulation/yolo_filtered_data/images/test',save=True)  # or img.jpg
```

```python
model_preds_train = 'runs/segment/predict2'
model_preds_test = 'runs/segment/predict2'
model_preds_val = 'runs/segment/predict2'


filter_folder_path = '/Users/vihaan/Workspace/Local_AI_Earth/DataManipulation/yolo_filtered_data'
filtered_images_folder_path_train = filter_folder_path + '/images/train'
filtered_label_folder_path_train = filter_folder_path + '/labels/train'
```

```python
############# TRAIN ###################

# List all files in the folder
all_files = os.listdir(filtered_images_folder_path_train)

# Filter out only JPG files
just_names = [os.path.splitext(filename)[0] for filename in all_files]

# Visualising three images side by side
plt.figure(figsize=(20,12))
f, axarr = plt.subplots(1,3,figsize=(20,12))

for i in range(len(just_names)):
    image = cv2.imread(filtered_images_folder_path_train + '/' + just_names[i] + '.jpeg')
    #plt.imshow(image)
    axarr[0].imshow(image)

    # Print the image with the labels
    image_path = filtered_images_folder_path_train + '/' + just_names[i] + '.jpeg'
    truth_labels_path = filtered_label_folder_path_train + '/' + just_names[i] + '.txt'
    axarr[1].imshow(data_utils.print_labels_on_image_and_return_image(image_path,truth_labels_path))

    # Printing the model predictions
    image = cv2.imread(model_preds_train + '/' + just_names[i] + '.jpeg')
    axarr[2].imshow(image)

    # Trying to save the image
    plt.savefig('combined_preds_truth/train/' + just_names[i] + '.jpeg')
    print(i)
```
