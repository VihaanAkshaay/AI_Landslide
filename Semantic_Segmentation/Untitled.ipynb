```python
# Visualising semantic segmentation loss

import numpy as np
```

```python
# Plotting the losses of AUNet

# Reading the numpy array from .pkl file:
val_losses = np.load('server_semantic_experiments/val_losses_aunet.pkl', allow_pickle=True)

train_losses = np.load('server_semantic_experiments/train_losses_aunet.pkl', allow_pickle=True)

# Plotting the losses:
import matplotlib.pyplot as plt

plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.show()
```
